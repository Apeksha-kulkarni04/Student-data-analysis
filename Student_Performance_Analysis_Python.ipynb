```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Study_Hours': [2, 4, 6, 8, 10, 3, 5, 7, 9, 1],
    'Attendance': [60, 70, 80, 90, 95, 65, 75, 85, 92, 55],
    'Previous_Score': [50, 55, 65, 75, 85, 58, 68, 78, 88, 45],
    'Final_Score': [52, 60, 70, 82, 90, 57, 68, 80, 89, 48]
}

df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.scatterplot(x='Study_Hours', y='Final_Score', data=df)
plt.title('Study Hours vs Final Score')
plt.show()

```

```python
sns.scatterplot(x='Attendance', y='Final_Score', data=df)
plt.title('Attendance vs Final Score')
plt.show()
```

```python
df.corr()
```

# Student Performance Analysis using Python

This project analyzes student academic performance data to identify key factors influencing final exam results using Python.


## Conclusion

The analysis shows that study hours have the strongest influence on students final scores, followed closely by attendance and previous academic performance. Students who study consistently and maintain good attendance tend to achieve better results. These insights can help educators focus on improving study habits and attendance to enhance overall student performance.
